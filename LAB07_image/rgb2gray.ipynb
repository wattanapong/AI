```python

import matplotlib.pyplot as plt
import cv2

image = cv2.imread("ict.jpg")
gray = image* [0.2989, 0.5870, 0.1140]

imgrgb = image[:,:,::-1]

fig, ax = plt.subplots(1, 3)
                                    
ax[0].imshow(imgrgb[:,:,0],cmap='gray')
ax[1].imshow(imgrgb[:,:,1],cmap='gray')
ax[2].imshow(imgrgb[:,:,2],cmap='gray')
# plt.savefig('ictgray.jpg')
cv2.imwrite('ictgray.jpg', image)
```

```python
import numpy as np

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

# Display the RGB image
ax1.imshow(imgrgb)
ax1.set_title('RGB Image')

# Display the histograms for each color channel
color = ('r', 'g', 'b')
for i, col in enumerate(color):
    histogram, bin_edges = np.histogram(imgrgb[:, :, i], bins=256, range=(0, 256))
    ax2.plot(bin_edges[0:-1], histogram, color=col, label=col.upper())

ax2.set_title('Color Histogram')
ax2.legend()
```
