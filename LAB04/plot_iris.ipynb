```python
import pandas
# iris = pandas.read_csv('IRIS/iris.data')
# iris = pandas.read_csv('IRIS/iris.data', header=None)
columns = ['sepal length','sepal width', 'petal length', 'petal width', 'class']
iris = pandas.read_csv('IRIS/iris.data', header=None, names=columns)

X = iris[['sepal length','sepal width', 'petal length', 'petal width']]
Y = iris['class']
```

```python
# plot sepal length between 3 classes
plt.figure(figsize=(10, 6))

# Plotting sepal length for each species
iris_classes = iris['class'].unique()
feature_name = 'petal width' # "petal length"  #'sepal width' # 'sepal length'
colors=[[1,0,0],[0,1,0],[0,0,1]]
for i, species in enumerate(iris_classes):
    species_data = iris[iris['class'] == species]
    plt.plot(species_data.index, species_data[feature_name], marker='o', linestyle='', color=colors[i], label=species)
```
