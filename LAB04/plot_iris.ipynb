```python
import pandas
# iris = pandas.read_csv('IRIS/iris.data')
# iris = pandas.read_csv('IRIS/iris.data', header=None)
columns = ['sepal length','sepal width', 'petal length', 'petal width', 'class']
iris = pandas.read_csv('IRIS/iris.data', header=None, names=columns)

X = iris[['sepal length','sepal width', 'petal length', 'petal width']]
Y = iris['class']
```

```python
# plot sepal length between 3 classes
plt.figure(figsize=(10, 6))

# Plotting sepal length for each species
iris_classes = iris['class'].unique()
feature_name = 'petal width' # "petal length"  #'sepal width' # 'sepal length'
colors=[[1,0,0],[0,1,0],[0,0,1]]
for i, species in enumerate(iris_classes):
    species_data = iris[iris['class'] == species]
    plt.plot(species_data.index, species_data[feature_name], marker='o', linestyle='', color=colors[i], label=species)
```

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons

# Generate moons dataset
X, y = make_moons(n_samples=100, noise=0.15, random_state=42)

# Plot the dataset
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Paired, edgecolors='k')
plt.title('Moons Dataset')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.colors as matcolor
from sklearn.datasets import load_wine

# Load the Wine dataset
wine = load_wine()
wine_data = wine.data
wine_target = wine.target
wine_feature_names = wine.feature_names
colors = ['']
# Choose two features for the scatter plot (e.g., the first two features)
fi1 = 0
fi2 = 1

# Plot the Wine dataset
plt.figure(figsize=(10, 6))
plt.scatter(wine_data[:, fi1], wine_data[:, fi2], c=wine_target, 
            cmap=plt.cm.Paired, edgecolor='k')
plt.xlabel(wine_feature_names[fi1])
plt.ylabel(wine_feature_names[fi2])
plt.title('Wine Dataset - Scatter Plot of Two Features')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Create some sample data for illustration
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)
y4 = np.exp(-0.1 * x) * np.sin(x)
y5 = x
y6 = x**2
y7 = np.sqrt(x)
y8 = np.log(x + 1)

# Create a 4x2 subplot layout
fig, axes = plt.subplots(nrows=4, ncols=2, figsize=(12, 10))

# Plot data on each subplot
axes[0, 0].plot(x, y1)
axes[0, 0].set_title('Feature 1')

axes[0, 1].plot(x, y2)
axes[0, 1].set_title('Feature 2')

axes[1, 0].plot(x, y3)
axes[1, 0].set_title('Feature 3')

axes[1, 1].plot(x, y4)
axes[1, 1].set_title('Feature 4')

axes[2, 0].plot(x, y5)
axes[2, 0].set_title('Feature 5')

axes[2, 1].plot(x, y6)
axes[2, 1].set_title('Feature 6')

axes[3, 0].plot(x, y7)
axes[3, 0].set_title('Feature 7')

axes[3, 1].plot(x, y8)
axes[3, 1].set_title('Feature 8')

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
```
