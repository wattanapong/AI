```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import accuracy_score
from sklearn import preprocessing
```

```python
train = pd.read_csv('StudentScore.Train.csv', header=None)
test = pd.read_csv('StudentScore.Test.csv', header=None)
```

```python
X_train, y_train = np.asarray(train.iloc[:,1]).reshape(-1,1), np.asarray(train.iloc[:,2]).reshape(-1,1)
X_test, y_test = np.asarray(test.iloc[:,1]).reshape(-1,1), np.asarray(test.iloc[:,2]).reshape(-1,1)
```

```python
# label_encoder = preprocessing.LabelEncoder()
# y_train = label_encoder.fit_transform(y_train)
# y_test = label_encoder.fit_transform(y_test)
```

```python
model = DecisionTreeClassifier()

model.fit(X_train, y_train)

y_pred = model.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)

print(f'Accuracy: {accuracy*100:.2f}%')
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(20,20))
plot_tree(model)
```
